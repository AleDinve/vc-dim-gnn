```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
datasets = ['NCI1', 'PTC_MR', 'PROTEINS']
act_list = ['atan']
mode = 'layers'
```

### Evolution in epochs of diff, varying the number of layers

```python

for dataset_name in datasets:
    for act in act_list:
        db = pd.read_csv('train1_23_11/all_data_'+dataset_name+'_'+ act + '_' + mode + '.csv')
        cols = db.columns 
        db = db[['epoch','diff','it','layer']]
        db = db[db['layer'].isin([2,4,6])]
        plot = sns.lineplot(data=db, x='epoch', y='diff', hue='layer')
        plot.figure.savefig('test_loss_23_11/loss_diff'+ dataset_name +'_'+act+ '_'+mode+'.png')
        plt.close()
```

### Evolution in epochs of diff, varying the hidden size

```python
for dataset_name in datasets:
    for act in act_list:
        db = pd.read_csv('data_new/all_data_'+dataset_name+'_'+act+'_layers')
        cols = db.columns 
        db = db[['epoch','diff','it','layer']]
        db = db[db['layer'].isin([16,32,64])]
        plot = sns.lineplot(data=db, x='epoch', y='diff', hue='layer')
        plot.figure.savefig('test_loss_23_11/loss_diff'+ dataset_name +'_'+act+ '_'+mode+'.png')
        plt.close()
```

### Evolution in hidden dimension of diff, varying certain epoch numbers

```python
for dataset_name in datasets:
    for act in act_list:
        db = pd.read_csv('data_new/all_data_'+dataset_name+ '_' + act +'_hd')
        cols = db.columns 
        db = db[['epoch','diff','it','hidden']]
        db = db[db['epoch'].isin([200,300,400])]
        plot = sns.lineplot(data=db, x='hidden', y='diff', hue='epoch')
        plot.figure.savefig('test_loss_23_11/loss_diff'+ dataset_name +'_'+act+'_x_'+mode+'.png')
        plt.close()
```
